From the file named 'sample.csv'. There are 2 series s1 and s2. you have to

Count how many times s1 cross above s2
Count how many times s1 cross below s2
Count how many times s1 cross s2 and detect at which date and time this happen and return a list of datetime.

```python
df3 = pd.read_csv('sample.csv')
```

```python
df3
```

```python
df3.rename(columns = {'0':'s1','1':'s2'},inplace = True)
```

```python
df3
```

Count how many times s1 cross above s2.

```python
cross_over = df3['s1'] >= df3['s2']
cross_above_dates = df3['Datetime'].loc[cross_over]
cross_above_count = cross_above_dates.count()
cross_above_list = cross_above_dates.values.tolist()
print('Count : ', cross_above_count)
print(cross_above_list)
```

Count how many times s1 cross below s2

```python
cross_down = df3['s1'] <= df3['s2']
cross_below_dates = df3['Datetime'].loc[cross_down]
cross_below_count = cross_below_dates.count()
cross_below_list = cross_below_dates.values.tolist()
print('Count : ', cross_below_count)
print(cross_below_list)
```

Count how many times s1 cross s2 and detect at which date and time this happen and return a list of datetime.

```python
crossing = ((df3['s1'] >= df3['s2']) | (df3['s1'] <= df3['s2']))
cross_dates = df3['Datetime'].loc[crossing]
print("Crossing Data count",cross_dates.count())
cross_dates_list = cross_dates.values.tolist()
print(cross_dates_list)
```
